## Проверка статистических гипотез

#### Немного про распределения

Распределение Бернулли $$
    X \~ B(p)
$$
Плотность вероятности: $$
    f(x|p) = p^x(1-p)^{1-x}(x)I_{x\in[0,1]}
$$
Функция-индикатор: $$
    I_{x\in[0,1]} = 1\ if\ x \in [0,1]\ else\ 0
$$
Матожидание: $$
    Ex = p
$$
Дисперсия: $$
    Dx = p(1-p)
$$

Биноминальное распределение $$
    X \~ Bin(n,p)
$$
Плотность вероятности: $$
    f(x|p) = C^{x}_np^x(1-p)^{1-x}(x)I_{x\in[0,1]}
$$
Число сочетаний: $$
    C^{x}_n = \frac{n!}{x!(n-x)!}
$$
Матожидание: $$
    Ex = np
$$
Дисперсия: $$
    Dx = np(1-p)
$$

Равномерное распределение $$
    X \~ U(a,b)
$$
Плотность вероятности: $$
    f(x) = I_{x\in[a,b]}(x)
$$
Матожидание: $$
    Ex = 0.5(b+a)
$$
Дисперсия: $$
    Dx = \frac{1}{3}(b^2+ab+a^2)
$$

### Что такое гипотеза?

Рассмотрим $$
    \hat{X} = (X_1, X_2 .. X_n)
$$
где Xn - случайные величины, одинаково распределенные с X.

Этот объект называется выборкой. 

У случайных величин в выборке одно распределение F. Предположим, что мы его не знаем.

Назовем гипотезой следующую конструкцию: 
$$
    H = \{F_i \in \hat{F}\}
$$

Критерий $$
    \delta: R \rightarrow \{H_1, H_2 ... H_n\}
$$

### Уровни статистической значимости

Уровень статистической значимости - фиксируемая вероятность совершения ошибки первого рода. Обычно используются 0.05, 0.01, 0.001.

### Алгоритм проверки статистической гипотезы по критическому значению

1. Задать уровень значимости;
2. Рассчитать критическое значение;
3. Посчитать статистику критерия;
4. Сравниваем с критическим значением.

### Алгоритм проверки статистической гипотезы по p-value

p-value - минимальное значение уровня значимости, при котором мы отвергаем нулевую гипотезу. Вероятность получить такие же или более экстремальные значения критерия в случае, если нулевая гипотеза верна.

1. Задать уровень значимости;
2. Посчитать статистику критерия;
3. Считаем достигнутый уровень значимости;
4. Сравниваем с вероятностью ошибки первого рода.

### Требования к критериям

1. Применимость;
2. Мощность критерия: $$
    P(H_1|H_1) = 1-\beta
$$
3. Состоятельность: $$
    lim_{n\rightarrow \infty} (1-\beta) = 1
$$
4. Несмещенность: $$
    1-\beta > \alpha\ if\ n \rightarrow \infty
$$
