```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.random.normal(size=100)
```

```python
plt.subplot(121)
plt.plot(a)
plt.subplot(122)
plt.plot(a-4)
plt.xlabel("Мягкость лапок")
plt.ylabel("Шелковистость шерстки")
plt.title("Характеристическая кривая котика")
plt.show()
```

```python
from matplotlib.patches import Patch
```

```python
plt.style.use("tableau-colorblind10")
fig, ax = plt.subplots(ncols=2)
#ax[0].plot(a, color="purple")
ax[0].fill_between(a, a-0.1, a+0.2, color=(0, 0, 0))
ax[0].plot(a-4)
ax[0].set_title("sfd")
ax[1].plot(a)
ax[1].plot(a-4)
ax[1].set_title("sfd")
ax[0].set_xlabel("324214")
ax[0].legend(
    loc="center",
    handles = [
        Patch(color="purple", label="gasgasfgasfg")
    ]
)
```

```python
plt.style.use("ggplot")
fig,ax = plt.subplots(nrows=1, ncols=3)
ax[0].plot(np.arange(10), np.array([0]*10))
ax[1].imshow(img)
ax[1].grid(False)
ax[2].plot(np.random.normal(size=100))
ax[1].arrow(x=250, y=800, dx=750, dy=20, color="red")
ax[1].annotate(xy=(250, 800), s="ТЫК", color="green")
```
