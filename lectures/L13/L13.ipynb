```python
from ludwig.api import *
```

```python
model_definition = {
    "input_features": [
        {"name": "age", "type": "numerical"},
        {"name": "sex", "type": "category"}
    ],
    "output_features": [
        {"name": "target", "type": "category"}
    ]
}
```

```python
model = LudwigModel(model_definition)
```

```python
import pandas as pd
```

```python
heart = pd.read_csv("heart.csv")
```

```python
train_stats = model.train(heart)
```

```python
model.save("kotik")
```
